ELO가 레이팅인것같다
- Id = indeX?
- groupId  = 그룹아이디
- matchId = 매치아이디
- assists = 어시스트 횟수(아군킬)
- boosts = 부스트 아이템 사용 횟수
- damageDealt = 피해량 (팀킬자해 제외)
- DBNOs = 기절횟수(Down but no outs)
- headshotKills = 헤드샷킬수
- heals = 회복템 사용횟수
- killPlace = 킬 횟수 순위
- killPoints = 킬 수로 산정한 ELO , 킬포가0이고 rankpoints값이 -1이 아니라면 무효
- `kills =` 총 킬 수
- killStreaks = 최대연속킬
- longestKill = 장거리 킬 거리
- matchDuration = 매치 시간 (초)
- matchType = 솔로 듀오 이런거
- maxPlace = 매치 내 최악의 등수
- numGroups = 게임 내 총 그룹 수
- rankPoints = 랭킹 ELO
- revives = 아군 부활
- rideDistance = 탈것이동거리
- roadKills = 로드킬 횟수
- swimDistance = 수영 거리
- teamKills = 팀킬 횟수
- vehicleDestroys = 차량 폭파 허;ㅅ스
- `walkDistance = ` 걸어간 거리
- `weaponsAcquired` = 얻은 무기 수
- `winPoints` = 승리횟수로 판단한 ELO
- `winPlacePerc` = 예측목표, 0부터 1까지 값, 1에 가까워질 수록 순위가 높아짐
