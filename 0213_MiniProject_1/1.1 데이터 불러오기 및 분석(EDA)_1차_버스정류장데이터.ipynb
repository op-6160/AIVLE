```python
# 아까 전에 네이버 지도로 확인했던 정류장ARS번호를 찾아볼까요?
# 종로2가사거리 = 01001
```

```python
# 데이터의 컬럼명을 바꿔볼까요? (to 버스정류장ARS번호)
seoul = seoul.rename(columns={'ARS-ID':'버스정류장ARS번호'})
```

```python
# 서울시 버스정류장 데이터와 bus_station을 합쳐볼까요? 
data = pd.merge(left=bus_station, right=seoul, how='inner', on='버스정류장ARS번호')
```

```python
# 구 별로 버스 정류장의 개수를 확인해 주세요
bus_station_ARS = data.groupby(by=["자치구"], as_index=False)["버스정류장ARS번호"].nunique()
bus_station_ARS
```

```python
# 버스정류장ARS번호 갯수를 확인해볼까요?
data['버스정류장ARS번호'].nunique()
```

#### [실습문제6] 구 별로 버스 노선이 몇 개나 지나가는지 궁금합니다.

* 구별로 버스 노선의 개수를 확인해 주세요

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
bus_station_line = data.groupby(by=["자치구"], as_index=False)["노선번호"].nunique()
bus_station_line
```

#### [실습문제7] 각 구별로 승차 총 승객수, 하차 총 승객수를 알아봅시다

* groupby 를 활용해 봅시다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
bus_station_sum = data.groupby('자치구',as_index=False)['승차총승객수','하차총승객수'].sum()
bus_station_sum
```

#### [실습문제8] 각 구별로 승차 평균 승객수, 하차 평균 승객수를 알아봅시다

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
bus_station_mean = data.groupby('자치구',as_index=False)['승차총승객수','하차총승객수'].mean()
bus_station_mean
```

#### [실습문제9] 데이터 프레임을 합쳐보겠습니다= '

```python
# 네 개 파일을 합쳐볼까요
# seoul_bus_station_ARS
# seoul_bus_station_line
# seoul_bus_station_sum
# seoul_bus_station_mean
t1 = pd.merge(left=bus_station_ARS,right=bus_station_line,how='inner',on='자치구')
t2 = pd.merge(left=t1,right=bus_station_sum,how='inner',on='자치구')
t3 = pd.merge(left=t2,right=bus_station_mean,how='inner',on='자치구')
t3.head()
```

```python
# column 이름을 바꿔주겠습니다.
t3.rename(columns={'승차총승객수_x':'승차총승객수','하차총승객수_x':'하차총승객수','승차총승객수_y':'승차평균승객수','하차총승객수_y':'하차평균승객수',
                   '노선번호':'노선수'}
                              ,inplace=True)
t3.head()
```

```python
# 해당 데이터프레임을 'df_seoul_bus_station.csv' 파일로 저장하세요.
t3.to_csv('./df_seoul_bus_station.csv')
```
