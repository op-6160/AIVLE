```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
#################
# train/test set 분리
# tesetsize= 10%
# 난수고정 2023
#################
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.1,random_state=2023)
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras
```

```python
#######################
# 어떤 전처리 작업 필요
# Hint : y.shape
#######################
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
########
# 모델링
########
keras.backend.clear_session()
model=keras.models.Sequential()

model.add( keras.layers.Input(shape=(4,)) )
model.add( keras.layers.Dense(32,activation='relu') )
model.add( keras.layers.Dense(16,activation='relu') )
model.add( keras.layers.Dense(3,activation='softmax') )

model.compile(loss='categorical_crossentropy',metrics=['acc'],optimizer='Adam')

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train,y_train,epochs=50)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
y_pred[:10], y_test[:10]
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import *
from sklearn.metrics import *
from tensorflow.keras.utils import *
```

```python
########
# 모델링
########
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2023,test_size=0.1)
```

```python
print('shape x,y ',x_train.shape,y_train.shape)
print('y_train unique',np.unique(y_train))
x_shape = x_train.shape[-1]
y_shape = len(np.unique(y_train))
dense_act = 'linear'
epochs = 50

if len(np.unique(y_train)) > 2:
    # when datset need one-hot-encoding
    y_train = to_categorical(y_train)
    y_test = to_categorical(y_test)
    dense_act = 'softmax'
    loss_func = 'categorical_crossentropy'

    print('y one-hot-encoded')
    print(' y unique',np.unique(y_train))

```

```python
###########
# 모델 학습
###########
# 1. clear session
keras.backend.clear_session()
# 2. model set
model = keras.models.Sequential()
# 3. layer set
model.add( keras.layers.Input(shape=x_shape) )
model.add( keras.layers.Dense(64,activation='relu') )
model.add( keras.layers.Dense(32,activation='relu') )
model.add( keras.layers.Dense(32,activation='relu') )
model.add( keras.layers.Dense(16,activation='relu') )
model.add( keras.layers.Dense(y_shape, activation=dense_act) )
# 4. compile model
model.compile(loss='categorical_crossentropy',metrics=['acc'],optimizer='adam')
model.summary()
```

```python
model.fit(x_train,y_train,epochs=epochs)
```

```python
###########
# 모델 예측
###########
y_pred = np.around(model.predict(x_test))
print(y_pred[:10])
print(y_test[:10])
```
